```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
```

```python
trends.build_payload(kw_list=["Metaverse"],)
data = trends.interest_by_region()
data = data.sort_values(by="Metaverse",ascending=False)
data = data.head(10)
print(data)
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Metaverse'])
data = data.interest_over_time()
```

```python
data.head
```

```python
fig, ax = plt.subplots(figsize=(10,8))
data['Metaverse'].plot()
plt.title('Total Google Searches for Metaverse', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
